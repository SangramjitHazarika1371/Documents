## Text Preprocessing

```python
# Apply Preprocessing on the Corpus

# Stop Words Removal
stop = set(stopwords.words('english'))

# Removal of Punctuation Marks
exclude = set(string.punctuation) 

# Lemmatization
lemma = WordNetLemmatizer() 

def clean(doc):
    
    # Convert text into lower case + split into words
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    
    # Remove any stop words present
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)  
    
    # Remove punctuations + normalize the text
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())  
    return normalized

# Clean data stored in a new list
clean_corpus = [clean(doc).split() for doc in corpus]  
```

```python
clean_corpus
```

## Creating Document Term Matrix

The advantage of using gensim for Document Term Matrix(DTM) is that we don't need to create the matrix from scratch explicitly. The gensim library has internal mechanism to create the Document Term Matrix(DTM).

The only requirement for gensis package is we need to pass the cleaned data in the form of tokenized words.

```python
# Creating the dictionary containing all the terms of our courpus(Specific to Genism syntax perspective)

dict_ = corpora.Dictionary(clean_corpus)

print(dict_)
```

```python
# Printing the terms
for i in dict_.values():
    print(i)
```

Now, the next step is to convert the corpus (the list of documents) into a document-term Matrix using the dictionary containing all the terms.

```python
# Converting list of documents (corpus) into Document Term Matrix using the dictionary 
doc_term_matrix = [dict_.doc2bow(i) for i in clean_corpus]
doc_term_matrix
```

The output implies:

    - Document wise we have the index of the word and its frequency.
    - The 0th word is repeated 1 time, then the 1st word repeated 1 and so on ...

## Implementation of LDA

```python
# Creating the object for LDA model using gensim library

Lda = gensim.models.ldamodel.LdaModel
```

```python
# Running and Training LDA model on the document term matrix.

ldamodel = Lda(doc_term_matrix, num_topics=6, id2word = dict_, passes=1, random_state=0, eval_every=None)
```

```python
# Print the topics with the indexes: 0,1,2, etc.

ldamodel.print_topics()
```

The output shows the dominance of a particular topic for the corresponding word.

## Extracting Topics from the Corpus

```python
print(ldamodel.print_topics(num_topics=6, num_words=5))

# num_topics mean: how many topics want to extract 
# num_words: the number of words that want per topic
```

## Assigning the topics to the documents

#### Pass the doc_term_matrix to the LDA model:

```python
# Printing the topic associations with the documents
count = 0
for i in ldamodel[doc_term_matrix]:
    print("Document: ",count+1,i)
    count += 1
```
